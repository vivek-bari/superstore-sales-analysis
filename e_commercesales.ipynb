```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('/Superstore_USA.csv')
dataset.head(5)
```

```python
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

```python
dataset['Order Priority']=dataset['Order Priority'].replace("Critical ",'Critical')
```

```python
dataset['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority",data=dataset)
plt.title("Order Priority acc to numbers")
plt.savefig('Order Priority.png')
plt.show()
```

```python
import pandas as pd

dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
dataset['Order Year'] = dataset['Order Date'].dt.year
```

```python
dataset['Order Year'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Year",data=dataset)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Year",data=dataset,hue='Customer Segment')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Order Year",y="Profit",data=dataset,estimator='sum')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x="Order Year",y="Profit",data=dataset,color='green')
plt.show()
```
